```python
from ucimlrepo import fetch_ucirepo

# fetch dataset
iris = fetch_ucirepo(id=53)

# data (as pandas dataframes)
X = iris.data.features
y = iris.data.targets

# metadata
print(iris.metadata)

# variable information
print(iris.variables)
```

```python
print(X)
```

```python
print(y);
```

```python
from sklearn.model_selection import train_test_split;
from sklearn.naive_bayes import GaussianNB;
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,ConfusionMatrixDisplay;
import seaborn as sms
import matplotlib.pyplot as plt
testsize=[];
acc_score=[];
for i in range(1,6):
  X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=i/10,random_state=20);
  print("Test size:-",i/10);
  classifier=GaussianNB();
  classifier.fit(X_train,Y_train.values.ravel());
  y_pred=classifier.predict(X_test);
  # cm=
  # (ConfusionMatrixDisplay(confusion_matrix=cm).plot())
  sms.heatmap(confusion_matrix(Y_test,y_pred),annot=True,cmap='Blues',fmt='d')
  plt.show()
  # print(cm);
  print("accuracy_score",accuracy_score(Y_test,y_pred))
  testsize.append(i/10);
  acc_score.append(accuracy_score(Y_test,y_pred));
  print("precision_score",precision_score(Y_test,y_pred,average='macro'))
  print("recall_score",recall_score(Y_test,y_pred,average='macro'))
  print("f1_score",f1_score(Y_test, y_pred, average='macro'))
  print("---------------------------------------------------------------")


plt.plot(testsize,acc_score);
plt.title('GaussianNB');
plt.xlabel('test percentage');
plt.ylabel('accuracy_score');
```

```python
from sklearn.model_selection import train_test_split;
from sklearn.tree import DecisionTreeClassifier

from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,ConfusionMatrixDisplay;
import seaborn as sms
import matplotlib.pyplot as plt
testsize=[];
acc_score=[];
for i in range(1,6):
  X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=i/10,random_state=20);
  print("Test size:-",i/10);
  classifier=DecisionTreeClassifier(criterion='entropy',random_state=0);
  classifier.fit(X_train,Y_train);
  y_pred=classifier.predict(X_test);
  # cm=
  # (ConfusionMatrixDisplay(confusion_matrix=cm).plot())
  sms.heatmap(confusion_matrix(Y_test,y_pred),annot=True,cmap='Blues',fmt='d')
  plt.show()
  # print(cm);
  print("accuracy_score",accuracy_score(Y_test,y_pred))
  testsize.append(i/10);
  acc_score.append(accuracy_score(Y_test,y_pred));
  print("precision_score",precision_score(Y_test,y_pred,average='macro'))
  print("recall_score",recall_score(Y_test,y_pred,average='macro'))
  print("f1_score",f1_score(Y_test, y_pred, average='macro'))
  print("---------------------------------------------------------------")


plt.plot(testsize,acc_score);
plt.title('Decisiontree(criterion=entropy)');
plt.xlabel('test percentage');
plt.ylabel('accuracy_score');
```

```python
from sklearn.model_selection import train_test_split;
from sklearn.tree import DecisionTreeClassifier;
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,ConfusionMatrixDisplay;
import seaborn as sms
import matplotlib.pyplot as plt
testsize=[];
acc_score=[];
for i in range(1,6):
  X_train,X_test,Y_train,Y_test=train_test_split(X,y,test_size=i/10,random_state=20);
  print("Test size:-",i/10);
  classifier=DecisionTreeClassifier(criterion='gini',random_state=0);
  classifier.fit(X_train,Y_train);
  y_pred=classifier.predict(X_test);
  # cm=
  # (ConfusionMatrixDisplay(confusion_matrix=cm).plot())
  sms.heatmap(confusion_matrix(Y_test,y_pred),annot=True,cmap='Blues',fmt='d')
  plt.show()
  # print(cm);
  print("accuracy_score",accuracy_score(Y_test,y_pred))
  testsize.append(i/10);
  acc_score.append(accuracy_score(Y_test,y_pred));
  print("precision_score",precision_score(Y_test,y_pred,average='macro'))
  print("recall_score",recall_score(Y_test,y_pred,average='macro'))
  print("f1_score",f1_score(Y_test, y_pred, average='macro'))
  print("---------------------------------------------------------------")


plt.plot(testsize,acc_score);
plt.title('Decisiontree(criterion=gini)');
plt.xlabel('test percentage');
plt.ylabel('accuracy_score');
```
